### Características de um Gráfico do Matplotlib

![Características de um Gráfico do Matplotlib](https://files.realpython.com/media/anatomy.7d033ebbfbc8.png)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.arange(1,5)

plt.plot(array)
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
```

```python
plt.plot(x, y)

plt.title('Seno')
plt.xlabel('x')
plt.ylabel('sen(x)')

plt.grid(True)
```
