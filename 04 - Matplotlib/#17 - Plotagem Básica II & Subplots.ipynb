4. Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0, 10, 0.01)

y1 = np.sin(x) # Primeira curva
y2 = np.cos(x) # Segunda curva
y3 = np.sin(x) + np.cos(3*x)
```

```python
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x) + cos(3x)')

plt.legend(loc='upper left')

plt.xticks(np.arange(0, 11))
plt.yticks([-2, 0, 2])

# plt.grid(True)
```

## Subplots

```python
plt.figure(figsize=(12,4))

plt.subplot(1, 3, 1) # Entrando na primeira região
plt.plot(x, y1)
plt.subplot(1, 3, 2) # Entrando na segunda região
plt.plot(x, y2)
plt.subplot(1, 3, 3)
plt.plot(x, y3)
plt.tight_layout()
```

```python
y4 = np.cos(x) + np.sin(3 * x)
```
